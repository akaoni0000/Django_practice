```python
import matplotlib.pyplot as plt

# データ
labels = ['a', 'b', 'c', '']  # 最後は0.0%用のラベル
sizes = [66.7, 16.7, 16.7, 0.0]
colors = ['#d62728', '#9467bd', '#8c564b', '#ffffff']  # 赤, 紫, 茶, 白（0%の部分）

# グラフ作成
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',
    startangle=90,
    counterclock=False,
    wedgeprops=dict(width=1.0)
)

# sizes：パーセンテージのデータ

# labels：ラベル

# colors：各セグメントの色

# autopct='%1.1f%%'：割合の表示（小数1位まで）

# startangle=90：最初のセグメントの開始角度（上から開始）

# counterclock=False：時計回りに描画（デフォルトは反時計回り）

# wedgeprops=dict(width=1.0)：セグメントの形状設定。width=1.0 で普通の円になります

# フォントサイズの調整
for text in texts:
    text.set_fontsize(12)
for autotext in autotexts:
    autotext.set_fontsize(10)

ax.axis('equal')  # 円を丸くする
plt.tight_layout()
plt.show()
```

```python
from collections import Counter
import fugashi

# コメントのリスト（例）
comments = [
    "このシステムは使いやすくて便利です。",
    "インターフェースが直感的で、操作しやすい。",
    "少し動作が重いと感じることがあります。",
    "サポートが丁寧で助かりました。",
    "UIがもう少し改善されると良いと思います。"
]

# 形態素解析器の準備
tagger = fugashi.Tagger()

# 名詞・形容詞・動詞を抽出
words = []
for comment in comments:
    for word in tagger(comment):
        pos = word.feature.pos1  # 品詞情報
        if pos in ['名詞', '形容詞', '動詞']:
            words.append(word.surface)

# 単語の頻度カウント
word_counts = Counter(words)

# 上位10件を表示
for i, (word, count) in enumerate(word_counts.most_common(10), start=1):
    print(f"{i}位: {word}（{count}回）")
```

```python
from collections import Counter
from janome.tokenizer import Tokenizer

# コメントのリスト（例）
comments = [
    "このシステムは使いやすくて便利です。",
    "インターフェースが直感的で、操作しやすい。",
    "少し動作が重いと感じることがあります。",
    "サポートが丁寧で助かりました。",
    "UIがもう少し改善されると良いと思います。"
]

# Tokenizerの準備
tokenizer = Tokenizer()

# 名詞・動詞・形容詞を抽出して原形でカウント
words = []

for comment in comments:
    for token in tokenizer.tokenize(comment):
        part_of_speech = token.part_of_speech.split(',')[0]  # 品詞（名詞、動詞など）
        base_form = token.base_form  # 原形（例：使う）

        if part_of_speech in ['名詞', '動詞', '形容詞']:
            if base_form != '*':  # base_form が "*" のものは除外
                words.append(base_form)

# 頻出単語のカウント
word_counts = Counter(words)

# 上位10件の単語を表示
for i, (word, count) in enumerate(word_counts.most_common(10), start=1):
    print(f"{i}位: {word}（{count}回）")
```

```python
from collections import Counter
from janome.tokenizer import Tokenizer

comments = [
    "このシステムは使いやすくて便利です。",
    "インターフェースが直感的で、操作しやすい。",
    "少し動作が重いと感じることがあります。",
    "サポートが丁寧で助かりました。",
    "UIがもう少し改善されると良いと思います。"
]

tokenizer = Tokenizer()
words = []

for comment in comments:
    for token in tokenizer.tokenize(comment):
        part_of_speech = token.part_of_speech.split(',')[0]
        base_form = token.base_form
        if part_of_speech in ['名詞', '動詞', '形容詞'] and base_form != '*':
            words.append(base_form)

word_counts = Counter(words)

# 同じ回数は同順位にする（スキップ付き）
prev_count = None
rank = 0
printed = 0

for word, count in word_counts.most_common():
    if count != prev_count:
        rank = printed + 1
    print(f"{rank}位: {word}（{count}回）")
    prev_count = count
    printed += 1
    if printed >= 10:  # 上位10件まで表示
        break
```

```python
import os
import json
from django.conf import settings

# samplesns/static/files の絶対パスを取得
files_base_dir = os.path.join(settings.BASE_DIR, 'samplesns', 'static', 'files')

# u1, u2 などのサブフォルダ名を取得
subdirs = [d for d in os.listdir(files_base_dir) if os.path.isdir(os.path.join(files_base_dir, d))]

all_data = []

for subdir in subdirs:
    subdir_path = os.path.join(files_base_dir, subdir)
    json_files = [f for f in os.listdir(subdir_path) if f.endswith('.json')]

    for json_file in json_files:
        file_path = os.path.join(subdir_path, json_file)
        with open(file_path, 'r', encoding='utf-8') as f:
            try:
                data = json.load(f)
                all_data.append({
                    "user": subdir,           # フォルダ名（例: u1, u2）
                    "filename": json_file,    # JSONファイル名
                    "content": data           # JSONの中身
                })
            except json.JSONDecodeError as e:
                print(f"読み込みエラー: {file_path} ({e})")

# 確認表示
for item in all_data:
    print(f"{item['user']}/{item['filename']}: {item['content']}")
```
